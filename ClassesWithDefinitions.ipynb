```python
from SPARQLWrapper import SPARQLWrapper, XML, JSON
import pandas as pd
```

```python
#hard code the source for the example
source = "http://sparql.hegroup.org/sparql/"
```

NOTE: the example uses The OBO IAO ontology annotation terms IAO_0000115 ("definition") and IAO_0000117 ("author"). If your ontology uses different annotation terms for definitions and authors, change them below.

```python
def classes_with_definitions(source):
    sparql = SPARQLWrapper(source)
    sparql.setReturnFormat(JSON)
    sparql.setQuery("""
    PREFIX obo-term: <http://purl.obolibrary.org/obo/>
    SELECT ?s ?label ?definition ?author
    FROM <http://purl.obolibrary.org/obo/merged/VO>
    {
    ?s a owl:Class .
    ?s rdfs:label ?label .
    ?s obo-term:IAO_0000115 ?definition .
    ?s obo-term:IAO_0000117 ?author .
    }
    """)
    try: 
        ret = sparql.queryAndConvert()
        results = []
        for r in ret["results"]["bindings"]:
            results.append(r)
        s = []
        label = []
        definition = []
        author = []
        for result in results:
            s.append(result["s"]["value"])
            label.append(result["label"]["value"])
            definition.append(result["definition"]["value"])
            author.append(result["author"]["value"])
        df = pd.DataFrame({"s": s, "label": label, "definition":definition, "author":author})
        pd.set_option('display.max_rows', None)
            
        return df 
    except:
        print("oh noes!")
```

```python
classes_with_definitions(source)
```
