```python
from SPARQLWrapper import SPARQLWrapper, XML, JSON
import pandas as pd
```

```python
#hard code the source for the example
source = "http://sparql.hegroup.org/sparql/"
```

To tailor the query, change the class specified in the where clause -- in the example it is obo-term:OAE_0000001, but you can make it any term you'd like - just make sure the relevant prefix is also in the query

```python
def find_all_subclasses(source):
    sparql = SPARQLWrapper(source)
    sparql.setReturnFormat(JSON)
    sparql.setQuery("""
    PREFIX obo-term: <http://purl.obolibrary.org/obo/>
    SELECT DISTINCT ?x ?label
    WHERE
    {
    ?x rdfs:subClassOf obo-term:OAE_0000001.
    ?x rdfs:label  ?label.
    }
    """)
    try: 
        ret = sparql.queryAndConvert()
        results = []
        for r in ret["results"]["bindings"]:
            results.append(r)
        x = []
        label = []
        for result in results: 
            label.append(result["label"]["value"])
            x.append(result["x"]["value"])
        df = pd.DataFrame({"label": label, "x": x})
        pd.set_option('display.max_rows', None)
        return df
    except: 
        print("Oh Noes!")
```

```python
find_all_subclasses(source)
```
