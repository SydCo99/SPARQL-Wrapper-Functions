```python
from SPARQLWrapper import SPARQLWrapper, XML, JSON
import pandas as pd
```

```python
#hard code the source 
source = "http://sparql.hegroup.org/sparql/"
```

NOTE: if you have a large ontology, this function will take a few minutes to run. Change the limit statement based on the size of your ontology - it is currently at 200 just so the query doesn't take too long. 

```python
def get_uris(source):
    sparql = SPARQLWrapper(source)
    sparql.setReturnFormat(JSON)
    sparql.setQuery("""
    PREFIX obo: <http://purl.obolibrary.org/obo/>
    PREFIX owl: <http://www.w3.org/2002/07/owl#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX oboInOwl: <http://www.geneontology.org/formats/oboInOwl#>
    PREFIX edam:<http://edamontology.org/>
    PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>

    SELECT DISTINCT ?entity ?label WHERE {

        ?entity a owl:Class .
        OPTIONAL  {?entity rdfs:label ?label .}

    }
    ORDER BY ?entity
    LIMIT 200
    """)
    try: 
        ret = sparql.queryAndConvert()
        results = []
        for r in ret["results"]["bindings"]:
            results.append(r)
        label = []
        entity = []
        for result in results:
            label.append(result["label"]["value"])
            entity.append(result["entity"]["value"])
        df = pd.DataFrame({"label": label, "entity": entity})
        pd.set_option('display.max_rows', None)
        
        return df 
    except:
        print("oh noes!")
```

```python
get_uris(source)
```
